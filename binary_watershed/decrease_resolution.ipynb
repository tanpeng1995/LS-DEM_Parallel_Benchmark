```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import tifffile as tif
from scipy.ndimage import zoom
%matplotlib inline
```

```python
imgs = tif.imread('/home/hasitha/Desktop/packed_img/gray_fabric_600.tif')
```

```python
imgs.shape
```

```python
# used for binary image
layers = imgs.shape[0]

# resize x and y
imgs2 = np.zeros((1200, 300, 300),dtype=bool)
for i in np.arange(layers):
    img = imgs[i]
    imgs2[i] = cv2.resize(np.float32(img),(300,300)).astype(np.bool)

# resize z
imgs3 = np.zeros((600, 300, 300),dtype=bool)
for i in np.arange(300):
    img = imgs2[:,:,i]
    imgs3[:,:,i] = cv2.resize(np.float32(img), (300,600)).astype(np.bool)
```

```python
# used for gray image
layers = imgs.shape[0]

# resize x and y
imgs2 = np.zeros((1200, 300, 300))
for i in np.arange(layers):
    img = imgs[i]
    imgs2[i] = cv2.resize(np.float32(img),(300,300))

# resize z
imgs3 = np.zeros((600, 300, 300))
for i in np.arange(300):
    img = imgs2[:,:,i]
    imgs3[:,:,i] = cv2.resize(np.float32(img), (300,600))
```
